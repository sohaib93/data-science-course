Start by creating 2 clusters. Then standardize the data and try again. Does it make a difference?

Use the Elbow rule to determine how many clusters are there.


## Import the relevant libraries

```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
from sklearn.cluster import KMeans
```

## Load the data

Load data from the csv file: <i> 'iris_dataset.csv'</i>.

```python
data = pd.read_csv('iris-dataset.csv')
data.head()
```

## Plot the data

For this exercise, try to cluster the iris flowers by the shape of their sepal. 

<i> Use the 'sepal_length' and 'sepal_width' variables.</i> 

```python
def plot_data(data_to_show,is_clustered):
    if(is_clustered == False):
        plt.scatter(data_to_show['sepal_length'],data_to_show['sepal_width'])
    else:
        plt.scatter(data_to_show['sepal_length'],data_to_show['sepal_width'], c=data_to_show['Cluster'], cmap='rainbow')
    plt.xlabel('sepal_length', fontsize=20)
    plt.xlabel('sepal_width', fontsize=20)
    plt.show()
    
plot_data(data,False)
```

# Clustering (unscaled data)

Separate the original data into 2 clusters.

```python
data_to_cluster = data.copy()
kmeans = KMeans(2)
kmeans.fit(data_to_cluster)
cluster_predictions = kmeans.fit_predict(data_to_cluster)
```

```python
cluster_predictions
```

```python
data['Cluster'] = cluster_predictions
plot_data(data,True)
```

# Standardize the variables

Import and use the <i> method </i> function from sklearn to standardize the data. 

```python
from sklearn import preprocessing

standardized_data = preprocessing.scale(data_to_cluster)
standardized_data
```

# Clustering (scaled data)

```python
kmeans = KMeans(2)
kmeans.fit(standardized_data)
```

```python
cluster_prediction_of_scaled_data = kmeans.fit_predict(standardized_data)
cluster_prediction_of_scaled_data
```

```python
data['Cluster'] = cluster_prediction_of_scaled_data
```

```python
plot_data(data,True)
```

## Take Advantage of the Elbow Method

### WCSS

```python
wcss = []
range_limit = 11
for i in range(1,range_limit):
    kmeans=KMeans(i)
    kmeans.fit(standardized_data)
    wcss.append(kmeans.inertia_)
    
wcss
```

### The Elbow Method

```python
number_clusters = range(1,range_limit)
plt.plot(number_clusters, wcss)
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('Within-cluster Sum of Squares')
```

```python
#from elbow method we select 2,3 and 5 numbers of clusters and will try them out below
#we already have 2 clustes selected above.
#lets try 3
```

```python
def get_cluster_predictions(std_data_to_cluster, k):
    kmeans = KMeans(k)
    kmeans.fit(std_data_to_cluster)
    predictions = kmeans.fit_predict(std_data_to_cluster)
    return predictions
```

```python
data['Cluster'] = get_cluster_predictions(standardized_data,3)
plot_data(data,True)
```

```python
#now lets try k = 5
```

```python
data['Cluster'] = get_cluster_predictions(standardized_data,5)
plot_data(data,True)
```

```python
#2 and 3 cluster colutions look reasonably better
```

# Compare your solutions to the original iris dataset
The original (full) iris data is located in iris_with_answers.csv. Load the csv, plot the data and compare it with your solution.

Obviously there are only 3 types, because that's the original (truthful) iris dataset.

The 2-cluster solution seemed good, but in real life the iris dataset has 3 SPECIES (a 3-cluster solution). Therefore, clustering cannot be trusted at all times. Sometimes it seems like x clusters are a good solution, but in real life, there are more (or less).

```python
real_data = pd.read_csv('iris-with-answers.csv')
real_data['species'].unique()
real_data['Cluster'] = real_data['species'].map({'setosa':0, 'versicolor':1 , 'virginica':2})
```

```python
#we can see that there are 3 species
```

```python
real_data.head()
```

```python
plot_data(real_data,True)
```

```python
plt.scatter(real_data['petal_length'],real_data['petal_width'], c=real_data['Cluster'], cmap='rainbow')
plt.title("ACTUAL RESULTS", fontsize=50)
plt.xlabel('petal_length', fontsize=20)
plt.xlabel('petal_width', fontsize=20)
plt.show()
```

```python
#petal info seems to cluster data better. so we plot our clustering solution with k = 3 based on petal info and see how close
#it is to actual
data['Cluster'] = get_cluster_predictions(standardized_data,3)
plt.scatter(data['petal_length'],data['petal_width'], c=data['Cluster'], cmap='rainbow')
plt.title("OUR CLUSTERING RESULTS", fontsize=50)
plt.xlabel('petal_length', fontsize=20)
plt.xlabel('petal_width', fontsize=20)
plt.show()
```
