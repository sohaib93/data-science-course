## Import the relevant libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
from sklearn.cluster import KMeans
```

## Load the data

Load data from the csv file: <i> 'Categorical.csv'</i>.

```python
raw_data = pd.read_csv('Categorical.csv')
raw_data.head()
```

```python
data = raw_data.copy()
data['continent'] = data['continent'].map({'North America':0,'Europe':1,'Asia':2,'Africa':3,'South America':4, 'Oceania':5,'Seven seas (open ocean)':6, 'Antarctica':7})
data.head()
```

## Select the features

```python
features = data.iloc[:,1:4]
features.head()
```

## Clustering

Use 4 clusters initially.

```python
kmeans = KMeans(7)
clusters = kmeans.fit_predict(features)
```

## Clustering results

```python
raw_data['Clusters'] = clusters
```

```python
raw_data
```

## Plot the data

```python
plt.scatter(raw_data['Longitude'],raw_data['Latitude'], c= raw_data['Clusters'], cmap='rainbow')
plt.xlabel('Longitude', fontsize=20)
plt.ylabel('Latitude', fontsize=20)
plt.show()
```
